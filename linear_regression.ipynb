```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5, 6, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85]  # Removed the extra element from y

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))
plt.scatter(x, y)  # Corrected the typo from 'scalter' to 'scatter'
plt.plot(x, mymodel)
plt.show()
```
